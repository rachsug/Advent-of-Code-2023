```python
# --- Day 2: Cube Conundrum ---
import os
from dotenv import load_dotenv
from aocd import get_data

def cube_conundrum():
    load_dotenv()
    data = get_data(session=os.environ.get('session'), day=2, year=2023)
    each_line = data.split('\n')

    keys = []
    for i in range(1, len(each_line)+1):
        keys.append('Game '+str(i))

    values = []
    for i in range(len(each_line)):
        values.append(each_line[i].replace(keys[i]+': ', '').replace(';', '').replace(',', ''))

    for value in range(len(values)):
        values[value] = values[value].split(' ')
    
        for i in range(len(values[value])):
            if i%2 == 0:
                values[value][i] = int(values[value][i])
    
    for j in range(len(values)):
        values[j] = [tuple(values[j][i: i+2]) for i in range(0, len(values[j]), 2)]
    
    games = set()
    for i in range(0, len(values)):
        for j in range(len(values[i])):
            if values[i][j][1] == 'red' and values[i][j][0] > 12:
                games.add(i+1)
            elif values[i][j][1] == 'green' and values[i][j][0] > 13:
                games.add(i+1)
            elif values[i][j][1] == 'blue' and values[i][j][0] > 14:
                games.add(i+1)

    hundred = 0
    for i in range(100+1):
        hundred += i
    return hundred-sum(games)    
        
cube_conundrum()
```
